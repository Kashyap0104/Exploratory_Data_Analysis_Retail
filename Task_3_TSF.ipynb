###Catplot is a relatively new addition to Seaborn that simplifies plotting that involves categorical variables. ... catplot function can do all these types of plots and one can specify the type of plot one needs with the kind parameter. The default kind in catplot() is “strip”, corresponding to stripplot().

```python
plt.figure(figsize=(10,7))
sns.catplot(x='Ship Mode', y='Profit',data=data.sort_values('Profit',ascending = False),kind="boxen", height = 6, aspect =3)
plt.title('ShipMode vs Profit',color='green')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.catplot(x='Category', y='Profit',data=data.sort_values('Profit',ascending = False),kind="boxen", height = 6, aspect =3)
plt.title('Category',color='green')
plt.show()
```

```python
df = data.groupby(['State'])[['Profit','Sales']].mean()
df
```

```python
df1 = data.groupby(['Category'])[['Discount','Profit','Sales']].mean()
df1
```

```python
df1.plot(kind='pie',subplots=True,figsize=(20,5),autopct='%0.2f%%')
plt.show()
```

## According to above pie chart
### **Technology** is having **highest** percentage of Sales and Profit. 
### **Furniture** is not having great Profit despite of having good percentage of Sales ane Discount.

```python
df2 = data.groupby(['Segment'])[['Profit','Sales','Discount']].mean()
df2
```

```python
df2.plot(kind='pie',figsize=(20,5),autopct='%0.2f%%',subplots=True)
```

## According to above pie chart
### **Home** **office** is having **Highest** percentege of  Profit and Sales.(Despite of having lowest Discount).
###**Consumer** Segment is having **lowest** percentage of Profit and Sales.
###**Corporate** segment is having **highest** percentage of Discount
## We can say that which is having highest percentage of Sales is getting highest percentage of Profit

```python
df3 = data.groupby(['City'])[['Profit']].mean()
df3
```
